```python
url = "https://kr.investing.com/stock-screener/Service/SearchStocks"
```

```python
payload = {"country[]" : 5, # 5번 미국... country[]는 나라 코드 
'sector' : "3,1,7,6,12,2,8,9,10,11,5,4",
'industry':'18,53,67,89,43,31,6,38,87,77,66,81,48,16,11,54,33,24,20,29,91,36,73,21,50,3,63,7,10,86,78,101,27,90,85,82,2,96,8,22,14,45,92,65,40,70,42,39,99,98,79,64,80,15,26,44,74,97,76,88,17,12,47,62,68,83,84,57,35,72,51,25,28,5,60,19,4,61,37,34,93,71,30,102,100,58,95,94,32,13,46,1,52,23,75,56,59,41,49,55,69,9',
'equityType':'ORD,DRC,Preferred,Unit,ClosedEnd,REIT,ELKS,OpenEnd,Right,ParticipationShare,CapitalSecurity,PerpetualCapitalSecurity,GuaranteeCertificate,IGC,Warrant,SeniorNote,Debenture,ETF,ADR,ETC,ETN',
'pn':'54', # 페이지 
'order[col]':'eq_market_cap',
'order[dir]':'d'          }
```

```python
r = requests.post(url, data=payload, headers= header)
```

```python
total = []
state = []
for idx in range(1, 220):
    print(idx)
    payload['pn'] = idx
    while(True):
        try:
            r = requests.post(url, data=payload, headers = header)
            rt = json.loads(r.text)
            rt['hits']
            cnt = 0
            break
        except KeyError:
            print('KeyError')
            print(r.status_code)
            if(cnt == 5):
                state = -1
                print("Keyerror exit")
                break
            time.sleep(3)
            cnt += 1
            
        except Exception as error:
            print(error)
            if(cnt == 5):
                state = -1
                print(error, 'Exit')
                break
            time.sleep(3)
            cnt += 1
    if(state == -1):
        print("Exit program")
        break
    for x in rt['hits']:
        total.append([x['name_trans'], x['stock_symbol'], x['exchange_trans'], x['industry_trans']])
```

```python
pd.DataFrame(total, columns=['name_trans', 'stock_symbol', 'exchage_trans', 'industry_trans'])
```

```python
us_stock = pd.DataFrame(total, columns=['name_trans', 'stock_symbol', 'exchange_trans', 'industry_trans'])
```

# DB저장

```python
import sqlite3
```

```python
con = sqlite3.connect('./common_stock.db')
```

```python
cur = con.cursor()
```

```python
query = """
        CREATE TABLE "us_stock" (
        'index' BIGINT NOT NULL
        "종목이름" TEXT NOT NULL,
        "종목코드" TEXT NOT NULL,
        "거래소" TEXT NOT NULL,
        "사업분야" TEXT NOT NULL,
        PRIMARY KEY('종목코드','index')
        )
"""
```

```python
con.execute(query)
```

```python
us_stock.to_sql("us_stock", con = con, if_exists='append', index= False)
```
