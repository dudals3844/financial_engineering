```python
import requests
import pandas as pd
import json
from bs4 import BeautifulSoup
```

```python
url = 'https://finance.naver.com/sise/sise_group.nhn?type=upjong'
```

```python
r = requests.get(url)
```

```python
bs = BeautifulSoup(r.text, 'html.parser')
```

```python
# upjong = bs.find(class_ = 'type_1').find_all("tr")
# # 오류남 왜?? --> table tag가 있어야 pandas가 인식
```

```python
upjong = bs.find("table", class_ = 'type_1')
```

```python
type(upjong)
```

```python
upjong_df = pd.read_html(str(upjong))[0]
```

```python
upjong_df
```

```python
upjong_df = upjong_df.dropna()
```

```python
upjong_df
```

```python
upjong_df.columns
```

## 콜롬에 값이 두줄에서 columns.droplevel을 이용해 한줄제거

```python
upjong_df.columns = upjong_df.columns.droplevel(0)
```
