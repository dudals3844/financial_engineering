```python
annual = []
quarterly = []
count = 0
for content in us_stock.iterrows():
    print (content[1]['stock_symbol'])
    name_code = content[1]['stock_symbol'].upper().replace("_", "-")
    # Anuual / Quarterly 가져오기 위해서 
    driver.get(url.format(name_code))
    for cnt in range(2):
        try:
            bs = BeautifulSoup(driver.page_source, 'lxml')
            bs_rt = bs.find("table", class_="Lh(1.7) W(100%) M(0)")
            df = pd.read_html(str(bs_rt))[0]
        except Exception as error:
            print (error)
            break
        
        df_trans = df.transpose()
        df_trans.columns = df_trans.iloc[0]
        df_trans.drop(df_trans.index[0], axis=0, inplace=True)
        df_trans['code'] = name_code
        if(cnt == 0):
            annual.append(df_trans)
            driver.find_element_by_css_selector('#Col1-1-Financials-Proxy > section > div.Mt\(18px\).Mb\(14px\) > div.Fl\(end\).smartphone_Mt\(4px\) > button > div > span').click()
        elif(cnt == 1):
            quarterly.append(df_trans)
    count+=1
    if(count == 20):
        break
```
