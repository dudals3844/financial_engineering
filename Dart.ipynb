```python
pdf_url = 'http://dart.fss.or.kr/pdf/download/pdf.do?'
popup_url = "http://dart.fss.or.kr/pdf/download/main.do?"
main_url = 'http://dart.fss.or.kr/dsaf001/main.do?'
html = {'eleId': ['0'],
 'offset': ['0'],
 'length': ['0'],
 'dtd': ['HTML']}
```

```python
for content in total_df.iterrows():
    print(content[1]['crp_nm'])
    url_result = requests.get(main_url, params={'rcpNo': content[1]['rcp_no']})
    pdf_download_re = re.findall(r"'(.*?)'",  BeautifulSoup(url_result.text, "html.parser").find("a", href="#download")['onclick'])[1]
    pdf_dict = {'rcp_no': content[1]['rcp_no'], 'dcm_no': pdf_download_re}
    
    while(True):
        try:
            r2 = requests.get(popup_url, params=pdf_dict)
            break
        except:
            print('Time error')
            time.sleep(3)
            pass
    html['rcpNo'] = pdf_dict['rcp_no']
    html['dcmNo'] = pdf_dict['dcm_no']
    if(BeautifulSoup(r2.text, 'html.parser').find("a")['href'].find("zip") > 0):
        download("http://dart.fss.or.kr/report/viewer.do?", html,  "./Dart/" + content[1]['crp_nm'] + "_" + content[1]['rcp_dt'] +
                "_" + content[1]['rpt_nm'].replace("/", "").replace(":", "") + ".html")
    else:
        download(pdf_url, pdf_dict,  "./Dart/" + content[1]['crp_nm'] + "_" + content[1]['rcp_dt'] +
                "_" + content[1]['rpt_nm'].replace("/", "").replace(":", "") + ".pdf")
    
```
