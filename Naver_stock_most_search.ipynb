```python
url = "https://finance.naver.com/sise/sise_rise.nhn"
```

```python
r = requests.get(url)
```

```python
bs = BeautifulSoup(r.text, 'lxml')
```

```python
item_result = bs.find("table", class_ = "item_list").find_all('input')
```

```python
item_result[0]['value']
```

```python
item = [x['value'] for x in item_result]
```

```python
item
```

```python
payload = {'menu': 'rise',
           'sosok': 0,
           'returnUrl': 'http://finance.naver.com/sise/sise_rise.nhn?sosok=0',
           'fieldIds': item[:7]
}
```

```python
r = requests.get(base_url, params=payload)
```

```python
bs = BeautifulSoup(r.text, 'html.parser')
```

```python
a = pd.read_html(str(bs.find("table", 'type_2')))[0]
```

```python
a
```

```python
rt_a = a.iloc[a['N'].dropna().index, :]
```
